```python
import pandas as pd

# 1. 데이터 불러오기
try:
    df = pd.read_excel('company_data_10000.xlsx')
    print("✅ [1/3] 데이터 로딩 완료")
    print("-" * 30)

    # 2. 데이터 구조 정보 확인 (info)
    print("✅ [2/3] 데이터 기본 정보 (df.info())")
    df.info()
    print("-" * 30)

    # 3. 데이터 상위 5개 행 확인 (head)
    print("✅ [3/3] 데이터 샘플 확인 (df.head())")
    # 주피터에서 표를 예쁘게 출력하기 위해 display() 함수를 사용합니다.
    display(df.head())

except Exception as e:
    print(f"❌ 에러가 발생했습니다: {e}")
```
