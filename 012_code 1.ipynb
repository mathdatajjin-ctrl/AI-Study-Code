```python
import pandas as pd

# 1. 데이터 불러오기
df = pd.read_excel('company_data_10000.xlsx')

# 2. 날짜 데이터 형식 변환 (날짜순 정렬을 위해 필수)
df['날짜'] = pd.to_datetime(df['월별 매출액'], format='%m월') # 1월, 2월 형식을 날짜로 인식

# 3. 최근 3개월 데이터 추출 (예: 10, 11, 12월 또는 마지막 3개 데이터)
# 전체 데이터 중 날짜가 가장 늦은 순서대로 3개를 가져옵니다.
recent_3_months = df.sort_values(by='날짜', ascending=False).head(3)

# 4. 분석에 필요한 주요 항목만 선택 (매출액, 방문자수, 마케팅비, 판매량)
analysis_table = recent_3_months[[
    '월별 매출액', 
    '지역별 매출',      # 매출액 항목
    '고객 방문 수',    # 방문자 수
    '마케팅 집행 내역', # 마케팅 관련
    '상품별 판매량'     # 대표상품 판매량
]]

display("=== [최근 3개월 주요 지표 분석 표] ===")
display(analysis_table)

# 5. (선택사항) 엑셀로 따로 저장하여 보고서 만들기
# analysis_table.to_excel('recent_3_months_report.xlsx', index=False)
```
