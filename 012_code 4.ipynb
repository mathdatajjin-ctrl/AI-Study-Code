```python
import pandas as pd

# 1. 원본 파일 불러오기
file_name = 'company_data_10000.csv'

try:
    # 한글 깨짐 방지를 위해 encoding='utf-8-sig' 사용
    df = pd.read_csv(file_name, encoding='utf-8-sig')
    print(f"✅ [1단계] '{file_name}' 파일을 성공적으로 불러왔습니다.")
    
    # 2. 데이터 확인 (원본 열 출력)
    print(f"📊 원본 데이터 열 목록: {df.columns.tolist()}")

    # 3. 데이터 전처리: 분석에 필요한 열만 선택
    # 삭제 대상: '개인정보_ID', '사무실_내선번호' 등 분석과 무관한 항목
    target_cols = ['직원 이름', '부서', '월별 매출액', '지역별 매출']
    df_processed = df[target_cols]
    
    print(f"✅ [2단계] 불필요한 열을 제거하고 {len(target_cols)}개의 핵심 열만 선택했습니다.")

    # 4. 결측치(비어있는 값) 처리
    # 데이터에 빈칸이 있다면 0으로 채우거나 삭제합니다.
    df_processed = df_processed.fillna(0)
    print("✅ [3단계] 결측치 확인 및 정제 완료")

    # 5. 전처리된 데이터를 새로운 파일로 저장
    output_name = 'final_analysis_data.csv'
    df_processed.to_csv(output_name, index=False, encoding='utf-8-sig')
    
    print("-" * 50)
    print(f"💾 [완료] 전처리가 끝난 파일이 '{output_name}'로 저장되었습니다.")
    
    # 6. 최종 결과 미리보기
    display(df_processed.head())

except FileNotFoundError:
    print("❌ 에러: 해당 파일이 폴더에 없습니다. 파일명을 확인해 주세요.")
except Exception as e:
    print(f"❌ 에러 발생: {e}")
```
