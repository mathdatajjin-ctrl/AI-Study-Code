```python
# 1. 분석에 필요한 라이브러리 불러오기 [cite: 133]
import pandas as pd

# 2. 전처리된 데이터 불러오기 (앞서 만든 파일 사용) [cite: 134]
df = pd.read_csv('public_air_quality.csv')

# 3. '특성(Feature)'과 '정답(Label)' 지정하기 [cite: 92, 93]
# 우리는 '미세먼지'와 '초미세먼지' 수치를 보고 -> '오존농도'를 예측하는 AI를 만든다고 가정합니다.

# AI가 학습할 재료 (특성들)
features = df[['미세먼지(PM10)', '초미세먼지(PM2.5)']] 

# AI가 맞혀야 할 목표 (정답)
label = df['오존농도']

print("▲ AI의 학습 재료 (Feature) 상단 5개:")
display(features.head())

print("\n▲ AI가 맞혀야 할 정답 (Label) 상단 5개:")
display(label.head())
```
