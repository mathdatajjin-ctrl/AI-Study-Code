```python
import pandas as pd

# 1. 파일 불러오기
file_name = 'company_data_10000.csv'

try:
    # 데이터 로드 (한글 깨짐 방지 옵션 포함)
    df = pd.read_csv(file_name, encoding='utf-8-sig')
    
    # [전처리 1] 분석에 불필요한 열 삭제 및 필요한 열만 선택
    # 원본 열 중 '개인정보_ID', '사무실_내선번호'를 제외한 핵심 열만 추출합니다.
    target_cols = ['직원 이름', '부서', '월별 매출액', '지역별 매출']
    df_processed = df[target_cols]
    
    # [전처리 2] 결측치 확인 (데이터가 비어있는 곳이 없는지 점검)
    null_count = df_processed.isnull().sum().sum()

    # 2. 결과 출력
    print(f"✅ 전처리 완료: 원본 {len(df.columns)}개 열 -> {len(df_processed.columns)}개 열로 정제")
    print(f"✅ 결측치 점검: 총 {null_count}개의 빈 데이터 발견")
    print("-" * 50)
    
    # 상위 5개 데이터 확인
    print("📊 정제된 데이터 상위 5행:")
    display(df_processed.head())

    # 3. 전처리된 데이터 별도 저장 (선택 사항)
    # df_processed.to_csv('final_processed_data.csv', index=False, encoding='utf-8-sig')

except FileNotFoundError:
    print("❌ 파일을 찾을 수 없습니다. 파일명이 'company_data_10000.csv'인지 확인해 주세요.")
```
